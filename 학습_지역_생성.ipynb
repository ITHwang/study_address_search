- '행정구역분류(개정_8자리)'의 타입을 int로 변경

```python
df['행정구역분류(개정_8자리)'] = df['행정구역분류(개정_8자리)'].fillna(0).astype('int64') 
df.dtypes
```

'행정구역명'과 '행정동(행정기관명)'의 값이 모든 행에서 같은지 확인하기

- '행정동(행정기관명)'은 행정구역명으로 대체하기
- 필요 없는 열 삭제

```python
drop_cols = [
    '행정동(행정기관명)',
    '행정구역분류(7자리)',
    '행정기관코드',
    '행정기관 생성일',
    '법정동 관할지역',
    '행정동 영문명칭',
    '비고',
]
df = df.drop(columns=drop_cols)
df.head()
```

'행정구역분류(개정_8자리)'가 8자리 미만인 지역은 모두 시군구 단위이므로 삭제(동읍면리 단위만 취급)

```python
df = df[df['행정구역분류(개정_8자리)'] // 10000000 != 0]
df.head()
```

법정동이 리 단위면 해당 읍을 찾아 제거(리를 가지고 있는 읍은 제외)

```python
for idx, row in df.iterrows():
    if row['법정동'][-1] == '리':
        eup_code = (row['법정동코드'] // 100) * 100
        eup_df = df[df['법정동코드'] == eup_code]
        if len(eup_df) == 0: continue
        df.drop(eup_df.index.tolist(), inplace=True)
```

null 체크

법정동(동읍면리) 기준으로 중심점을 찾으려고함.

행정동 안에 여러 법정동이 있는 경우에는 상관없지만,

법정동 안에 여러 행정동이 있는 경우에는 각기 다른 행정동들이 하나의 중심점에 의존하게 되는 한계를 지님.

중심점까지 구하고 테이블 병합(매핑) 후 리 단위 지역을 가지고 있는 면 단위 지역은 행 삭제하기(이미 리 단위의 세부 지역 중심점이 존재함)

```python
df_emd = pd.read_csv('읍면동_중심점.csv')
df_emd
```

```python
df_emd.dtypes
```

```python
df_emd = df_emd[['법정동코드', '중심점_x', '중심점_y']]
df_emd
```

```python
df_emd.dtypes
```

```python
df_li = pd.read_csv('리_중심점.csv')
df_li = df_li[['법정동코드', '중심점_x', '중심점_y']]
df_li
```

```python
df_li.dtypes
```

법정동코드를 키로 잡고 병합해주기

읍면동의 법정동코드는 뒤에 0이 두 개 더 붙어야 함.

```python
df_emd["법정동코드"] = df_emd['법정동코드'] * 100
df_emd
```

원래 df와 읍면동 중심점 df 병합

```python
df_emd_merge = pd.merge(df, df_emd, how='left')
df_emd_merge
```

원래 df와 리 중심점 df 병합

```python
df_li_merge = pd.merge(df, df_li, how='left')
df_li_merge
```

```python
df_emd_merge['중심점_x'].isnull().sum()
```

```python
df_li_merge['중심점_x'].isnull().sum()
```

리 단위 지역 중심점을 병합한 df에서 널 값이 있는 코드를 df_emd에서 찾기

```python
def find_coor_from_emd(addr_code):
    df = df_emd_merge[df_emd_merge['법정동코드'] == addr_code]
    if len(df) == 0: raise Exception(f'읍면동 테이블에서 해당 법정동 코드({addr_code})가 없습니다.')
    return df.iloc[0]['중심점_x'], df.iloc[0]['중심점_y']
    
for idx, row in df_li_merge.iterrows():
    if np.isnan(row['중심점_x']):
        addr_code = row['법정동코드']
        x, y = find_coor_from_emd(addr_code)
        df_li_merge.loc[idx, '중심점_x'] = x
        df_li_merge.loc[idx, '중심점_y'] = y
```

통계청 통계분류포털상 존재하지만, 행정안전부 도로명주소 db에 존재하지 않는 법정동(두 기관, shp 파일을 공유하는 사이트에서 업데이트가 느린 것일 수도)

```python
df_nan = df_li_merge[df_li_merge['중심점_x'].isnull()]
df_nan
```

통합 엑셀, 중심점이 있는 지역 엑셀, 중심점이 없는 엑셀 저장

```python
df_all = df_li_merge
df_notnull = df_li_merge[df_li_merge['중심점_x'].notnull()]
df_null = df_li_merge[df_li_merge['중심점_x'].isnull()]
```

```python
df_all.to_excel('전체_학습_지역.xlsx', index=False)
df_notnull.to_excel('가능_학습_지역.xlsx', index=False)
df_null.to_excel('불가능_학습_지역.xlsx', index=False)
```
