```python
import numpy as np
import pandas as pd
from geopy import distance
from pyproj import Transformer
```

## 1. 데이터 로드 

```python
df_town = pd.read_csv('town.csv')
df_town.head()
```

## 2. 각 동네마다 반경 거리별 근처 동네 개수 구하기

```python
def init_obj(*distances):
    return {f'{dist}km': 0 for dist in distances}
```

```python
towns = []
distances = [3, 5, 7]
transformer = Transformer.from_crs("epsg:5179", "epsg:4326")

for i, row in df_town.iterrows():
    if i % 20 != 0: continue # 약 1000개 동네들 샘플링
    
    obj = init_obj(*distances)
    src_point = transformer.transform(row['중심점_y'], row['중심점_x'])
    for inner_i, inner_row in df_town.iterrows():
        if inner_i == i: continue
        dst_point = transformer.transform(inner_row['중심점_y'], inner_row['중심점_x'])
        try:
            dist_between = distance.distance(src_point, dst_point).km
            for dist in distances:
                if dist_between <= dist:
                    obj[f'{dist}km'] += 1
        except Exception as e: print(e) 
    towns.append(obj)
    if i % 100 == 0: print(f'{i} / {len(df_town)} 완료')
```

```python
import pickle
with open('near_town_count.pickle', 'wb') as f:
    pickle.dump(towns, f, pickle.HIGHEST_PROTOCOL)
```
