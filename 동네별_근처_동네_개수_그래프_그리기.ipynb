```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
with open('near_town_count.pickle', 'rb') as f:
    towns = pickle.load(f)
print(towns)
```

```python
cnt = {}
for town in towns:
    for key in town.keys():
        if key not in cnt: cnt[key] = []
        cnt[key].append(town[key])
```

```python
len(cnt['3km']), len(cnt['5km']), len(cnt['7km'])
```

```python
plt.hist(cnt['3km'], bins=50)
plt.savefig('3km.png')
```

```python
plt.hist(cnt['5km'], bins=50)
plt.savefig('5km.png')
```

```python
plt.hist(cnt['7km'], bins=50)
plt.savefig('7km.png')
```
